```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
heart_data = pd.read_csv("/content/heartdisease.csv")
```

```python
heart_data.head(7)
```

```python
isnull = heart_data.isnull().sum()
print(isnull)
```

```python
heart_data.shape
```

```python
#corralation matrix

crm = heart_data.corr()
plt.figure(figsize=(11,11))
sns.heatmap(crm, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Heart Disease Data")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
x = heart_data.drop('target', axis=1)
y = heart_data['target']
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y , test_size=0.2, random_state=42)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
print("TESTING X",x_train.shape)
print("TESTING Y",x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
knn_model.fit(x_train, y_train)
```

```python
##accuracy score
knn_train_predict = knn_model.predict(x_train)
knn_test_predict = knn_model.predict(x_test)
accuracy_train = accuracy_score(y_train, knn_train_predict)
accuracy_test = accuracy_score(y_test, knn_test_predict)
print("Training Accuracy:", accuracy_train)
print("Testing Accuracy:", accuracy_test)
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear')
```

```python
svm_model.fit(x_train,y_train)
```

```python
train_predict = svm_model.predict(x_train)
accuracy_train = accuracy_score(y_train,train_predict)
print("Training ACCURACY SCORE = ",accuracy_train)
```

```python
test_predict = svm_model.predict(x_test)
accuracy_test = accuracy_score(y_test,test_predict)
print("TESTING ACCURACY SCORE = ",accuracy_test)
```
